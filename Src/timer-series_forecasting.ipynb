```python
# Stationarity check function
def check_stationarity(series):
    result = adfuller(series, autolag='AIC')
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'\t{key}: {value}')
# Check stationarity for 'Spot_Price_72'
check_stationarity(data['Spot_Price_72'])
```

#### checking the stationarity of a time series dataset, specifically using the Augmented Dickey-Fuller (ADF) test. 

1. **Defining the `check_stationarity` Function**:
   - This function takes a time series (a sequence of data points in time order) as input.
   - It performs the Augmented Dickey-Fuller test on this series, which is a statistical test used to check for stationarity in a time series.
   - The function prints out the ADF statistic, the p-value, and the critical values for different confidence levels (1%, 5%, and 10%).

2. **Understanding the ADF Test**:
   - **ADF Statistic**: This is the result of the test statistic. More negative values indicate stronger rejection of the null hypothesis (which states that the series has a unit root, i.e., it is non-stationary).
   - **p-value**: This tells you the probability of observing the data assuming the null hypothesis is true. A lower p-value (typically < 0.05) suggests that you can reject the null hypothesis, indicating the series is stationary.
   - **Critical Values**: These values correspond to different confidence levels. If the ADF statistic is more negative than the critical value, the null hypothesis can be rejected at that confidence level.

3. **Applying the Function to `Spot_Price_72`**:
   - The function `check_stationarity` is called with the 'Spot_Price_72' series from your data.
   - The output shows that the ADF statistic is -1.444517, and the p-value is 0.560758. The ADF statistic is not more negative than any of the critical values.
   - Based on these results, the p-value is higher than common significance levels (0.05, 0.01), meaning there's not enough evidence to reject the null hypothesis. Therefore, the 'Spot_Price_72' series is likely non-stationary according to this test.

In time series analysis, particularly when using models like ARIMA, it's important to work with stationary data. Stationarity implies that the statistical properties of the series (like mean and variance) do not change over time. Non-stationary data often require transformations, such as differencing, to make them stationary before modeling.

```python
# Differencing the series
data['Spot_Price_72_diff'] = data['Spot_Price_72'].diff().dropna()
```

### The line `data['Spot_Price_72_diff'] = data['Spot_Price_72'].diff().dropna()` is used for differencing the 'Spot_Price_72' time series data. Let's break down what this does:

1. **Differencing**: 
   - The `.diff()` method is used to calculate the difference between each consecutive value in the 'Spot_Price_72' series. This is a common technique in time series analysis, particularly for making a non-stationary time series stationary.
   - Differencing helps to remove trends and seasonality from a time series. By subtracting the previous value from the current value, the transformed series can often reveal a more stationary process, which is an assumption of many time series models, including ARIMA.

2. **Drop NA**:
   - After differencing, the first value in the series becomes `NaN` (since there's no preceding value to subtract from the first value). The `.dropna()` method is used to remove any `NaN` values that result from the differencing process.

3. **Creating a New Column**:
   - The differenced series is stored in a new column in the dataframe named 'Spot_Price_72_diff'. This allows for the preservation of the original series while providing a transformed version for further analysis.

In summary, this line of code is preparing the 'Spot_Price_72' time series for further analysis, particularly for use in statistical models that require stationarity, by creating a differenced version of the series.

Stationarity is an important assumption in many time series models, including ARIMA models, because it implies that the statistical properties of the series (mean, variance, autocorrelation, etc.) are constant over time, which is key for reliable forecasting and analysis. With a stationary series, you can proceed with further time series modeling or analysis.

The Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) plots are instrumental in identifying the appropriate lag values for the autoregressive (AR) and moving average (MA) components in ARIMA modeling. Here's how they are typically used:

1. **ACF Plot**:
   - The ACF plot shows the correlation of the time series with its own lagged values.
   - For ARIMA models, the number of MA terms \( q \) can be initially identified by the lag after which the ACF plot cuts off and the bars become statistically insignificant (i.e., within the blue shaded area representing the confidence interval).

2. **PACF Plot**:
   - The PACF plot shows the partial correlation of the time series with its own lagged values, controlling for the values of the time series at all shorter lags.
   - The number of AR terms \( p \) in ARIMA models is often suggested by the lag value at which the PACF plot cuts off and becomes insignificant.

From your ACF plot:
- There is a gradual decline in the correlation values, and they remain positive and significant for many lags. This could suggest a higher order of MA terms \( q \), as the correlations are slowly decreasing rather than cutting off sharply.

From your PACF plot:
- There is a significant spike at lag 1, and then the PACF values drop off and generally stay within the confidence interval. This indicates that you might consider \( p = 1 \) for the AR term, as the correlation is significant at the first lag and cuts off afterward.

So, based on these plots, a possible starting model could be ARIMA(1, d, Q) where \( d \) is the degree of differencing you have already determined is necessary to make your series stationary (which was 1 in your case), and \( Q \) would need further examination of the ACF plot to determine the exact number of MA terms, potentially considering more than 2 given the slow decay in the ACF. However, the ACF's slow decline might also be indicative of a data series that is not yet fully stationary or has a long memory, which could mean that an ARIMA model may not be the best choice without additional differencing or transformation.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Dense

# Filter data for the year 2000 onwards
data = data[data['Year'] >= 2000]

# One-hot encode the 'Country' column
df_encoded = pd.get_dummies(data, columns=['Country'])

# Separate features and target variable
features = df_encoded.drop(columns=['Spot_Price_72'])
target = df_encoded['Spot_Price_72']

# Perform train-test split
split_ratio = 0.8
split_index = int(len(df_encoded) * split_ratio)
train_features, test_features = features[:split_index], features[split_index:]
train_target, test_target = target[:split_index], target[split_index:]

# Scale the features
feature_scaler = MinMaxScaler()
train_features = feature_scaler.fit_transform(train_features)
test_features = feature_scaler.transform(test_features)

# Scale the target variable separately
target_scaler = MinMaxScaler()
train_target = target_scaler.fit_transform(train_target.values.reshape(-1, 1))
test_target = target_scaler.transform(test_target.values.reshape(-1, 1))

def create_sequences(features, target, sequence_length):
    sequences, labels, indices = [], [], []
    for i in range(len(features) - sequence_length):
        sequences.append(features[i:i + sequence_length])
        labels.append(target[i + sequence_length])
        indices.append(i + sequence_length)  # The index of the label in the original dataset
    return np.array(sequences), np.array(labels), np.array(indices)

# Create sequences and labels
sequence_length = 10
X_train, y_train, train_indices = create_sequences(train_features, train_target, sequence_length)
X_test, y_test, test_indices = create_sequences(test_features, test_target, sequence_length)

# Determine the number of features
num_features = X_train.shape[2]

# Build and compile the model
model = Sequential()
model.add(Bidirectional(LSTM(64, activation='relu'), input_shape=(sequence_length, num_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
batch_size = 64
epochs = 50
model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, verbose=1)

# Predict and inverse transform
y_pred = model.predict(X_test)
y_pred = target_scaler.inverse_transform(y_pred)
y_test = target_scaler.inverse_transform(y_test)

# Reshape y_test and y_pred for easier manipulation
y_test = y_test.reshape(-1)
y_pred = y_pred.reshape(-1)

# Country-wise MSE calculation
country_columns = [col for col in df_encoded.columns if col.startswith('Country_')]
country_mse = {}
original_test_features = df_encoded[split_index:]  # Original test features before scaling

for country in country_columns:
    country_mask = original_test_features[country].iloc[test_indices] == 1  # Align with test_indices
    if country_mask.any():  # Check if there's any data for this country
        country_y_test = y_test[country_mask]
        country_y_pred = y_pred[country_mask]
        mse = mean_squared_error(country_y_test, country_y_pred)
        country_mse[country.replace('Country_', '')] = mse

# Display MSE for each country
print("Mean Squared Error (MSE) by Country:")
for country, mse in country_mse.items():
    print(f"{country}: {mse}")

# Calculate overall MSE
mse = mean_squared_error(y_test, y_pred)
print(f"Overall Mean Squared Error (MSE): {mse}")

# Visualization
plt.plot(y_test, label='Actual')
plt.plot(y_pred, label='Predicted', color='red')
plt.legend()
plt.show()

# Prepare the latest data for each country to predict for 2023
latest_sequences = {}
for country in df_encoded.columns[df_encoded.columns.str.startswith('Country_')]:
    # Filter data for the last sequence_length rows for each country
    country_data = df_encoded[df_encoded[country] == 1].tail(sequence_length)
    if len(country_data) == sequence_length:  # Ensure enough data is available
        latest_sequence = feature_scaler.transform(country_data.drop(columns=['Spot_Price_72']))
        latest_sequences[country] = latest_sequence

# Predict Spot_Price_72 for 2023
predictions_2023 = {}
for country, sequence in latest_sequences.items():
    sequence = sequence.reshape(1, sequence_length, num_features)  # Reshape for model input
    predicted_price = model.predict(sequence)
    predicted_price = target_scaler.inverse_transform(predicted_price)  # Inverse transform the prediction
    predictions_2023[country.replace('Country_', '')] = predicted_price[0][0]

# Display predictions for 2023
print("Forecasted Spot_Price_72 for 2023 by Country:")
for country, prediction in predictions_2023.items():
    print(f"{country}: {prediction}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Dense, Conv1D, MaxPooling1D, Flatten
from tensorflow.keras.regularizers import l1, l2

# Filter data for the year 2000 onwards
data = data[data['Year'] >= 2000]

# One-hot encode the 'Country' column
df_encoded = pd.get_dummies(data, columns=['Country'])

# Separate features and target variable
features = df_encoded.drop(columns=['Spot_Price_72'])
target = df_encoded['Spot_Price_72']

# Perform train-test split
split_ratio = 0.8
split_index = int(len(df_encoded) * split_ratio)
train_features, test_features = features[:split_index], features[split_index:]
train_target, test_target = target[:split_index], target[split_index:]

# Scale the features
feature_scaler = MinMaxScaler()
train_features = feature_scaler.fit_transform(train_features)
test_features = feature_scaler.transform(test_features)

# Scale the target variable separately
target_scaler = MinMaxScaler()
train_target = target_scaler.fit_transform(train_target.values.reshape(-1, 1))
test_target = target_scaler.transform(test_target.values.reshape(-1, 1))

def create_sequences(features, target, sequence_length):
    sequences, labels, indices = [], [], []
    for i in range(len(features) - sequence_length):
        sequences.append(features[i:i + sequence_length])
        labels.append(target[i + sequence_length])
        indices.append(i + sequence_length)  # The index of the label in the original dataset
    return np.array(sequences), np.array(labels), np.array(indices)

# Create sequences and labels
sequence_length = 10
X_train, y_train, train_indices = create_sequences(train_features, train_target, sequence_length)
X_test, y_test, test_indices = create_sequences(test_features, test_target, sequence_length)

# Determine the number of features
num_features = X_train.shape[2]

# Build and compile the 1D CNN model
model = Sequential()
model.add(Conv1D(64, kernel_size=3, activation='relu', input_shape=(sequence_length, num_features), kernel_regularizer=l2(0.01)))
model.add(MaxPooling1D(pool_size=2))
model.add(Conv1D(128, kernel_size=3, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(50, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
batch_size = 64
epochs = 100
model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, verbose=1)


# Predict and inverse transform
y_pred = model.predict(X_test)
y_pred = target_scaler.inverse_transform(y_pred)
y_test = target_scaler.inverse_transform(y_test)

# Reshape y_test and y_pred for easier manipulation
y_test = y_test.reshape(-1)
y_pred = y_pred.reshape(-1)

# Country-wise MSE calculation
country_columns = [col for col in df_encoded.columns if col.startswith('Country_')]
country_mse = {}
original_test_features = df_encoded[split_index:]  # Original test features before scaling

for country in country_columns:
    country_mask = original_test_features[country].iloc[test_indices] == 1  # Align with test_indices
    if country_mask.any():  # Check if there's any data for this country
        country_y_test = y_test[country_mask]
        country_y_pred = y_pred[country_mask]
        mse = mean_squared_error(country_y_test, country_y_pred)
        country_mse[country.replace('Country_', '')] = mse

# Display MSE for each country
print("Mean Squared Error (MSE) by Country:")
for country, mse in country_mse.items():
    print(f"{country}: {mse}")

# Calculate overall MSE
mse = mean_squared_error(y_test, y_pred)
print(f"Overall Mean Squared Error (MSE): {mse}")

# Visualization
plt.plot(y_test, label='Actual')
plt.plot(y_pred, label='Predicted', color='red')
plt.legend()
plt.show()


# Prepare the latest data for each country to predict for 2023
latest_sequences = {}
for country in df_encoded.columns[df_encoded.columns.str.startswith('Country_')]:
    # Filter data for the last sequence_length rows for each country
    country_data = df_encoded[df_encoded[country] == 1].tail(sequence_length)
    if len(country_data) == sequence_length:  # Ensure enough data is available
        latest_sequence = feature_scaler.transform(country_data.drop(columns=['Spot_Price_72']))
        latest_sequences[country] = latest_sequence

# Predict Spot_Price_72 for 2023
predictions_2023 = {}
for country, sequence in latest_sequences.items():
    sequence = sequence.reshape(1, sequence_length, num_features)  # Reshape for model input
    predicted_price = model.predict(sequence)
    predicted_price = target_scaler.inverse_transform(predicted_price)  # Inverse transform the prediction
    predictions_2023[country.replace('Country_', '')] = predicted_price[0][0]

# Display predictions for 2023
print("Forecasted Spot_Price_72 for 2023 by Country:")
for country, prediction in predictions_2023.items():
    print(f"{country}: {prediction}")
```
