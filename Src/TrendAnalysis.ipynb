```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the merged dataset
file_path = 'merged_dataset.csv'
main_df = pd.read_csv(file_path)
```

```python
# Convert all columns to float where possible, except 'Country' and 'Year'
for col in main_df.columns:
    if col not in ['Country', 'Year']:
        main_df[col] = pd.to_numeric(main_df[col], errors='coerce')
```

```python
# Group by 'Year' and calculate mean for each column
yearly_means = main_df.groupby('Year').mean()
```

```python
# Plotting trends of the mean of each column versus 'Year'
for column in yearly_means.columns:
    plt.figure(figsize=(10, 5))
    plt.plot(yearly_means.index, yearly_means[column], marker='o', linestyle='-')
    plt.title(f'Yearly Mean Trend of {column}')
    plt.xlabel('Year')
    plt.ylabel(f'Mean of {column}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
