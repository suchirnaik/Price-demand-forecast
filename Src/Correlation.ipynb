```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load the merged dataset
file_path = 'merged_dataset.csv'
main_df = pd.read_csv(file_path)
```

```python
# Convert all columns to float where possible, except 'Country' and 'Year'
for col in main_df.columns:
    if col not in ['Country', 'Year']:
        main_df[col] = pd.to_numeric(main_df[col], errors='coerce')

# Group by 'Year' and calculate mean for each column
yearly_means = main_df.groupby('Year').mean()
```

```python
# Calculate the correlation matrix for the yearly means
correlation_matrix = yearly_means.corr()
```

```python
correlation_matrix
```

```python
# Create a directory for saving plots if it doesn't exist
plot_dir = 'Trend_Analysis_Plots'
os.makedirs(plot_dir, exist_ok=True)

# Plotting and saving the correlation matrix heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()

# Save the heatmap
heatmap_file_path = os.path.join(plot_dir, 'correlation_heatmap.png')
plt.savefig(heatmap_file_path)
plt.close()

# Return the path of the saved heatmap
heatmap_file_path
```
